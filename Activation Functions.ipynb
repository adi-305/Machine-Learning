    <h1>Important Points about the Sigmoid Function</h1>
    <ol>
        <li><strong>Activation Function:</strong> The sigmoid function is widely used as an activation function in neural networks, especially in binary classification problems.</li>
        <li><strong>Range:</strong> The output of the sigmoid function ranges between 0 and 1, making it ideal for models that predict probabilities.</li>
        <li><strong>Formula:</strong> The sigmoid function is mathematically represented as: 
            <br><code>Ïƒ(x) = 1 / (1 + e<sup>-x</sup>)</code></li>
        <li><strong>Smooth Gradient:</strong> The function provides a smooth and differentiable gradient, which is important for backpropagation in neural networks.</li>
        <li><strong>Vanishing Gradient Problem:</strong> Sigmoid functions can cause the vanishing gradient problem for very high or very low input values, slowing down the learning process in deep networks.</li>
    </ol>

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))

def dsegmoid(z):
    return sigmoid(z)*(1 - sigmoid(z))

z = np.linspace(-10,10, num = 100)
g = sigmoid(z)
dg = dsegmoid(z)
plt.figure(figsize = (6, 4))
plt.plot(z, g, 'r')
plt.plot(z, dg, 'b--')
plt.legend(['Sigmoid', 'Derivative of Sigmoid'])
plt.grid()
plt.show()
```

  <h1>Important Points about the ReLU Function</h1>
    <ol>
        <li><strong>Activation Function:</strong> ReLU is one of the most commonly used activation functions in deep learning models, especially for hidden layers.</li>
        <li><strong>Formula:</strong> The ReLU function is defined as:
            <br><code>ReLU(x) = max(0, x)</code></li>
        <li><strong>Sparsity:</strong> ReLU introduces sparsity by outputting zero for negative input values, which helps reduce the number of active neurons in a network.</li>
        <li><strong>Efficient Computation:</strong> ReLU is computationally efficient due to its simple mathematical operation, leading to faster training and inference.</li>
        <li><strong>Issue with Dying Neurons:</strong> ReLU can suffer from the "dying ReLU" problem, where neurons can get stuck and output zero for all inputs if their weights are not updated properly.<
</body>
</html>


```python
def relu(z):
    return np.maximum(z,0)

def drelu(z):
    return z > 0

z = np.linspace(-10,10, num = 100)
g = relu(z)
dg = drelu(z)
plt.figure(figsize = (6, 4))
plt.plot(z, g, 'r')
plt.plot(z, dg, 'b--')
plt.legend(['RELU', 'Derivative of RELU'])
plt.grid()
plt.show()
```

<body>
    <h1>Important Points about the Leaky ReLU Function</h1>
    <ol>
        <li><strong>Activation Function:</strong> Leaky ReLU is a variant of the ReLU activation function, designed to allow small negative values to flow, solving the issue of dying neurons.</li>
        <li><strong>Formula:</strong> The Leaky ReLU function is defined as:
            <br><code>Leaky ReLU(x) = x (if x &gt; 0) or Î±x (if x â‰¤ 0)</code> where Î± is a small constant (typically 0.01).</li>
        <li><strong>Non-Zero Gradient for Negative Values:</strong> Unlike ReLU, Leaky ReLU assigns a small slope (Î±) for negative inputs, preventing neurons from completely dying.</li>
        <li><strong>Smooth Training:</strong> The small negative slope ensures that gradient-based optimization methods can update the weights even for inputs that would have been zero in regular ReLU.</li>
        <li><strong>Variation in Hyperparameter Î±:</strong> The constant Î± can be adjusted, typically between 0.01 and 0.1, depending on the task, to control how much negative information is allowed through.</li>
    </ol>
</body>
</html>


```python
def leaky_relu(z, alpha=0.1):
    return np.where(z > 0, z, alpha * z)

def d_leaky_relu(z, alpha=0.1):
    return np.where(z > 0, 1, alpha)

z = np.linspace(-10,10, num = 100)
g = leaky_relu(z)
dg = d_leaky_relu(z)
plt.figure(figsize = (6, 4))
plt.plot(z, g, 'r')
plt.plot(z, dg, 'b--')
plt.legend(['Leaky RELU', 'Derivative of Leaky RELU'])
plt.grid()
plt.show()
    
```

<body>
    <h1>Important Points about the Tanh Function</h1>
    <ol>
        <li><strong>Activation Function:</strong> The Tanh function is commonly used as an activation function in neural networks, particularly for hidden layers.</li>
        <li><strong>Range:</strong> The Tanh function outputs values in the range of <code>-1</code> to <code>1</code>, which is useful for centering data.</li>
        <li><strong>Formula:</strong> The Tanh function is mathematically represented as:
            <br><code>tanh(x) = (e<sup>x</sup> - e<sup>-x</sup>) / (e<sup>x</sup> + e<sup>-x</sup>)</code></li>
        <li><strong>Centered at Zero:</strong> Unlike the Sigmoid function, Tanh is zero-centered, which helps the model converge faster because the mean of the activations will be close to zero.</li>
        <li><strong>Vanishing Gradient Problem:</strong> Similar to Sigmoid, Tanh can also cause the vanishing gradient problem for very large or very small inputs, which can slow down learning in deep networks.</li>
    </ol>
</body>
</html>


```python
def tanh(z):
    return (np.exp(z) - np.exp(-z))/(np.exp(z) + np.exp(-z))

def d_tanh(z):
    return 1 - tanh(z)**2

z = np.linspace(-10, 10, num=100)
g = tanh(z)
dg = d_tanh(z)

plt.figure(figsize=(6, 4))
plt.plot(z, g, 'r')
plt.plot(z, dg, 'b--')
plt.legend(['tanh', 'Derivative of tanh'])
plt.grid()
plt.show()
```

<body>
    <h1>Important Points about the Softmax Function</h1>
    <ol>
        <li><strong>Activation Function:</strong> The Softmax function is often used in the final layer of a neural network for multi-class classification tasks.</li>
        <li><strong>Probability Distribution:</strong> It converts the output scores (logits) into a probability distribution, where the sum of all probabilities equals 1.</li>
        <li><strong>Formula:</strong> The Softmax function is defined as:
            <br><code>Softmax(x<sub>i</sub>) = e<sup>x<sub>i</sub></sup> / Î£ e<sup>x<sub>j</sub></sup></code> for each output <code>x<sub>i</sub></code>.</li>
        <li><strong>Multi-Class Predictions:</strong> Softmax is ideal for making predictions over multiple classes, as it assigns a probability to each class, making it easier to interpret the results.</li>
        <li><strong>Exponential Scaling:</strong> The exponential function in Softmax scales the input logits, making the larger scores more prominent while pushing smaller values closer to zero, enhancing the separation between predicted classes.</li>
    </ol>
</body>
</html>


```python
def softmax(z):
    exp_z = np.exp(z - np.max(z))  # Subtract max for numerical stability
    return exp_z / exp_z.sum(axis=0)

def d_softmax(z):
    s = softmax(z)
    return s * (1 - s)  # Simplified element-wise derivative

z = np.linspace(-5, 5, num=100)
g = softmax(z)
dg = d_softmax(z)

plt.figure(figsize=(6, 4))
plt.plot(z, g, 'r')
plt.plot(z, dg, 'b--')
plt.legend(['Softmax', 'Derivative of Softmax'])
plt.grid()
plt.show()
```
